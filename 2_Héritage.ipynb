# **Héritage:**
- Du point de vue de la programmation, cela signifie généralement "l'héritage ou le transfert de caractéristiques d'une classe parent à une classe enfant sans aucune modification".
- La nouvelle classe est appelée classe dérivée/enfant  et celle dont elle est dérivée est appelée classe parent/base  .

image.png

# **Héritage unique :**
- L'héritage à un seul niveau permet à une classe dérivée d'hériter des caractéristiques d'une classe parent unique.

-  On  prends la classe parente et on crée un constructeur (__init__), la classe elle-même initialise les attributs avec des paramètres ('name', 'age' et 'salary').

```python
#This is a parent class
class employee1():
    def __init__(self, name, age, salary):  
        self.name = name
        self.age = age
        self.salary = salary
```

- Création d'une classe enfant 'childemployee' qui hérite des propriétés d'une classe parent et enfin instanciation des objets 'emp1' et 'emp2' par rapport aux paramètres.

```python
#This is a child class
class childemployee(employee1):
    def __init__(self, name, age, salary,id):
        self.name = name
        self.age = age
        self.salary = salary
        self.id = id
```

- Enfin, on a imprimé l'âge de emp1. Eh bien, vous pouvez faire un tas de choses comme imprimer tout le dictionnaire ou le nom ou le salaire.

 

```python
emp1 = employee1('harshit',22,1000)
print(emp1.age)
```

# **Héritage à plusieurs niveaux :**
- L'héritage à plusieurs niveaux permet à une classe dérivée d'hériter des propriétés d'une classe parent immédiate qui, à son tour, hérite des propriétés de sa classe parent.

- défini la superclasse comme Employee

```python
class employee():#Super class
    def __init__(self,name,age,salary):  
        self.name = name
        self.age = age
        self.salary = salary
```

-  La classe enfant comme childemployee1

```python
class childemployee1(employee):#First child class
    def __init__(self,name,age,salary):
        self.name = name
        self.age = age
        self.salary = salary
```

-  Maintenant, childemployee1 agit en tant que parent pour childemployee2.


```python
class childemployee2(childemployee1):#Second child class
    def __init__(self, name, age, salary):
        self.name = name
        self.age = age
        self.salary = salary
```

- Instancié deux objets 'emp1' et 'emp2' où je passe les paramètres "name", "age", "salary" pour emp1 de la superclasse "employee" et "name", "age, "salary" et "id » de la classe parente « childemployee1 »

```python
emp1 = employee('harshit',22,1000)
emp2 = childemployee1('arjun',23,2000)
 
print(emp1.age)
print(emp2.age)
```

# **Héritage hiérarchique :**
- L'héritage de niveau hiérarchique permet à plusieurs classes dérivées d'hériter des propriétés d'une classe parent.

```python
class employee():
    def __init__(self, name, age, salary): #Hierarchical Inheritance
        self.name = name
        self.age = age
        self.salary = salary
```

- vous pouvez clairement voir qu'il existe deux classes enfants "childemployee1" et "childemployee2".
- Ils héritent des fonctionnalités d'une classe mère commune qui est « employee ».


```python
class childemployee1(employee):
    def __init__(self,name,age,salary):
        self.name = name
        self.age = age
        self.salary = salary
```

```python
class childemployee2(employee):
    def __init__(self, name, age, salary):
        self.name = name
        self.age = age
        self.salary = salary
```

- Les objets 'emp1' et 'emp2' sont instanciés par rapport aux paramètres 'name', 'age', 'salary'.

```python
emp1 = employee('harshit',22,1000)
emp2 = employee('arjun',23,2000)
print(emp1.age)
print(emp2.age)
```

# **Héritage multiple :**
- L'héritage à plusieurs niveaux permet à une classe dérivée d'hériter des propriétés de plusieurs classes de base.

```python
class employee1():#Parent class
    def __init__(self, name, age, salary):  
        self.name = name
        self.age = age
        self.salary = salary
```

```python
class employee2():#Parent class
    def __init__(self,name,age,salary,id):
     self.name = name
     self.age = age
     self.salary = salary
     self.id = id
```

- classe enfant "childemployee", qui hérite des deux classes parentes employee1 et employee2

```python
class childemployee(employee1,employee2):
    def __init__(self, name, age, salary,id):
     self.name = name
     self.age = age
     self.salary = salary
     self.id = id
```

- Instanciant les objets 'emp1' et 'emp2' par rapport aux paramètres des classes parentes.

```python
emp1 = employee1('harshit',22,1000)
emp2 = employee2('arjun',23,2000,1234)
print(emp1.age)
print(emp2.id)
```
