# **Les bases du décorateur**
- Un décorateur est un appelable qui fonctionne comme un wrapper, prenant une fonction comme argument et effectuant un traitement avec la fonction décorée.

# **Pizza exemple with python decorators**

image.png

- Nous avons des fonctions imbriquées adding_cheese et inner_function .
- La fonction adding_cheese prend une fonction comme argument et renvoie fonction_intérieure .
- Dans inner_function , il exécute les instructions d'impression avant et après l'appel de la fonction passée en argument à added_cheese .
- L'ajout de @adding_cheese au-dessus de la fonction baking_pizza décore la cuisson de la pizza avec l'ajout de fromage en d'autres termes , l'ajout de fromage est le décorateur et la cuisson de la pizza est la fonction décorée .

```python
def adding_cheese(func):
    def inner_function():
        print('Just adding more cheese here...')
        func()
        print('And Done!')
    return inner_function

@adding_cheese
def baking_pizza():
    print('Just baking pizza here...')
```

```python
baking_pizza()
```

- Le décorateur adding_toppings ajoute des fonctions décorées à list toppings .  

- Dans la fonction principale, nous retournons le contenu des garnitures de liste et appelons les fonctions adding_olives , adding_onions , adding_tomatoes et adding_apple .
- La fonction adding_toppings retourne quelle fonction est décorée puis ajoutée à la liste des garnitures .

```python
toppings = list()

def adding_toppings(func):
    print(f"Adding topping {func}...")
    toppings.append(func)
    return func

@adding_toppings
def adding_olives():
    print('Added Olives...')
    
@adding_toppings
def adding_onions():
    print('Added Onions...')
    
@adding_toppings
def adding_tomatoes():
    print('Added tomatoes...')
    
def adding_apple():
    print('This is getting out of hand now. GET OUT!!!')
```

```python
print('Stated adding toppings')
print(f'Toppings added so far {toppings}')
adding_olives()
adding_onions()
adding_tomatoes()
adding_apple()
```

# **Décorateur dans une classe**
- Considérons quand le décorateur est défini à l'intérieur d'une classe,

```python
class pizza():
    @classmethod
    def adding_cheese(cls, func):
        def inner_function(*args, **kwargs):
            print('Just adding more cheese here...')
            ret_val = func(*args, **kwargs)
            return ret_val
        return inner_function
    
@pizza.adding_cheese
def baking_pizza(name=None):
    print(f'{name} baking pizza here...')
    return "And Done!"
```

- La classe pizza a une fonction décoratrice adding_cheese qui décore la fonction baking_pizza .
- Dans ce cas, nous décorons avec @pizza.adding_cheese .

```python
ret_val = baking_pizza(name="Blah Blah")
print(ret_val)
```
